###### Converting the ActivityDate into Date format and extracting the Month and day of week 

```python
df['ActivityDate']=pd.to_datetime(df['ActivityDate'])
df['weekday']=df['ActivityDate'].dt.dayofweek
df['month']=df['ActivityDate'].dt.month
```

```python
df.head()
```

###### Dropping Activity Date and Tracker Distance 

```python
df.drop('ActivityDate',inplace=True,axis=1)
df.drop('TrackerDistance',inplace=True,axis=1)
```
